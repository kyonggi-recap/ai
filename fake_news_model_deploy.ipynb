```python
import boto3

s3 = boto3.client('s3')
bucket_name = 'my-rag-model-bucket'

s3.create_bucket(
    Bucket=bucket_name,
    CreateBucketConfiguration={
        'LocationConstraint': 'ap-northeast-2'  # 서울이면 이렇게
    }
)
```

```python
from sagemaker.pytorch import PyTorchModel
from sagemaker import get_execution_role, Session

role = get_execution_role()
sagemaker_session = Session()

model = PyTorchModel(
    entry_point="inference.py",
    model_data="s3://my-rag-model-bucket/model/model.tar.gz",
    role=role,
    framework_version="1.13",
    py_version="py39"
)

predictor = model.deploy(
    instance_type="ml.m5.large",
    initial_instance_count=1,
    endpoint_name="fake-true"  # 원하는 이름으로 고정!
)
```

```python
import numpy as np

# 입력: 배치 크기 1, query+3 context (4), 임베딩 차원 1536
example_input = np.random.rand(1, 4, 1536).tolist()  # float32 값 리스트로 변환
result = predictor.predict(example_input)
print("예측 결과:", result)
```
